```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
import scipy.signal
import scipy.misc
```

```python
from google.colab import widgets

def compare_images(imgs, figsize=(10,10), **kwargs):
  labels = ['image ' + str(i) for i in range(len(imgs))]
  tb = widgets.TabBar(labels, location='top')
  for i, img in enumerate(imgs):
    with tb.output_to(i, select=(i == 0)):
      plt.figure(figsize=figsize)
      plt.imshow(img, **kwargs)
```

```python
img1 = scipy.misc.ascent() / 255.0
img2 = img1 + np.random.normal(size=img1.shape) * 0.1
img3 = scipy.ndimage.gaussian_filter(img2, (2.0, 2.0))

compare_images((img1, img2, img3), vmin=0, vmax=1)
```
